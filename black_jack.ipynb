## **åˆ¶ä½œè€…ã®ã‚³ãƒ¡ãƒ³ãƒˆ**

é€šå¸¸ã®ãƒ–ãƒ©ãƒƒã‚¯ã‚¸ãƒ£ãƒƒã‚¯ã§ã¯ã€ãƒ‡ã‚£ãƒ¼ãƒ©ãƒ¼ã¯ã€ã‚«ãƒ¼ãƒ‰ã®åˆè¨ˆãŒ16ä»¥ä¸‹ã®æ™‚ã¯å¿…ãšãƒ’ãƒƒãƒˆã€17ä»¥ä¸Šã§ã‚ã‚Œã°å¿…ãšã‚¹ã‚¿ãƒ³ãƒ‰ã¨ã„ã†ãƒ«ãƒ¼ãƒ«ãŒã‚ã‚Šã¾ã™ã€‚ã—ã‹ã—ã€ç§ãŒä½œã£ãŸã‚²ãƒ¼ãƒ ã§ã¯ã€ãƒ—ãƒ¬ã‚¤ãƒ¤ãƒ¼ã‚ˆã‚Šç‚¹æ•°ãŒä½ã„å ´åˆã€21æœªæº€ãªã‚‰ãƒ‡ã‚£ãƒ¼ãƒ©ãƒ¼ã‚‚ã‚«ãƒ¼ãƒ‰ã‚’å¼•ã„ã¦é€†è»¢ã§ãã‚‹ã‚ˆã†ã«ãªã£ã¦ã„ã‚‹ã®ã§ã€å‹ã¡ã«ãããªã£ã¦ã„ã¾ã™ã€‚

ã‚³ãƒ¼ãƒ‰ã‚’ä½œã‚‹ã†ãˆã§é›£ã—ã„ã¨æ„Ÿã˜ãŸã¨ã“ã‚ã¯ã€ã‚«ãƒ¼ãƒ‰ã®å‰²ã‚ŠæŒ¯ã‚Šã§ã™ã€‚åå‰ã‚’ä»˜ã‘ã‚‹ã“ã¨ã§åˆ†é¡ã—ã€ã‚ã‹ã‚Šã‚„ã™ãã—ã¾ã—ãŸã€‚ã¾ãŸã€è¦‹ãŸã¨ãã«ã‚³ãƒ¼ãƒ‰ã®æ„å‘³ãŒåˆ†ã‹ã‚‹ã‚ˆã†ã«ã‚„ã‚„ã“ã—ã„ã¨ã“ã‚ã‚„æ–°ã—ãçŸ¥ã£ãŸé–¢æ•°ã«ã¯ã‚³ãƒ¡ãƒ³ãƒˆã‚’ä»˜ã‘ã¾ã—ãŸã€‚

æ¬¡ä½œã‚‹ã¨ãã¯æœ¬å½“ã®ãƒ–ãƒ©ãƒƒã‚¯ã‚¸ãƒ£ãƒƒã‚¯ã®ã‚ˆã†ã«ãƒãƒƒãƒ—ã‚’è¿½åŠ ã—ã¦ã€è³­ã‘ã‚’ã§ããŸã‚‰é¢ç™½ããªã‚Šãã†ã ã¨æ€ã£ãŸã®ã§ã€æŒ‘æˆ¦ã—ã¦ã¿ãŸã„ã§ã™ã€‚

## **ä½œæˆæ™‚é–“**

ä½œæˆæ™‚é–“ï¼š**ç´„12æ™‚é–“**

```python
import random
import itertools

# ===== ã‚«ãƒ¼ãƒ‰ã®æº–å‚™ =====
suit = list('â™ â™¥â™¦â™£')
rank = ['A'] + list(map(str, range(2, 11))) + list('JQK')
cards = list(map(lambda x: x[0] + x[1], itertools.product(suit, rank)))

deck = cards.copy()  #ã‚³ãƒ”ãƒ¼ã—ãªã„ã¨ã€ãƒªã‚¹ãƒˆã€Œcardsã€ãŒå¤‰ã‚ã£ã¦ã—ã¾ã†
random.shuffle(deck)

# ===== ã‚«ãƒ¼ãƒ‰ã®ç‚¹æ•°è¨ˆç®— =====
def card_value(card):  #card_valueé–¢æ•°ï¼šæ¸¡ã•ã‚ŒãŸã‚«ãƒ¼ãƒ‰ 1 æšãŒä½•ç‚¹ã«ãªã‚‹ã‹ã‚’åˆ¤æ–­ã™ã‚‹ãŸã‚ã®é–¢æ•°
    rank = card[1:]  #æ–‡å­—åˆ—ã®2æ–‡å­—ç›®ä»¥é™ã‚’åˆ‡ã‚Šå‡ºã™
    if rank in ['J', 'Q', 'K']:
        return 10
    elif rank == 'A':
        return 11
    else:
        return int(rank)  #æ•°ã‚’ãã®ã¾ã¾è¿”ã™

def calculate_score(hand):  #calculate_scoreé–¢æ•°ï¼šãƒ—ãƒ¬ã‚¤ãƒ¤ãƒ¼ã‚„ãƒ‡ã‚£ãƒ¼ãƒ©ãƒ¼ã®æ‰‹æœ­ï¼ˆã‚«ãƒ¼ãƒ‰ï¼‰ã‹ã‚‰åˆè¨ˆç‚¹æ•°ã‚’è¨ˆç®—ã™ã‚‹é–¢æ•°
    score = 0  # ç‚¹æ•°ã®åˆè¨ˆã‚’0ã‹ã‚‰ã‚¹ã‚¿ãƒ¼ãƒˆ
    ace_count = 0  # Aã®æšæ•°ã‚’ã‚«ã‚¦ãƒ³ãƒˆ

    # 1æšãšã¤æ‰‹æœ­ã‚’è¦‹ã¦ã„ã
    for card in hand:
        value = card_value(card)  # ã‚«ãƒ¼ãƒ‰ã®ç‚¹æ•°ã‚’èª¿ã¹ã‚‹
        score += value  # ç‚¹æ•°ã«åŠ ãˆã‚‹

        # ã‚‚ã—Aã ã£ãŸã‚‰ã€Aã®æ•°ã‚’ã‚«ã‚¦ãƒ³ãƒˆ
        if card[1:] == 'A':
            ace_count += 1

    # ç‚¹æ•°ãŒ21ã‚’è¶…ãˆã¦ã„ãŸã‚‰ã€Aã®å€¤ã‚’11â†’1ã«å¤‰ãˆã¦èª¿æ•´ã™ã‚‹
    while score > 21 and ace_count > 0:
        score -= 10
        ace_count -= 1

    return score

# ===== ã‚²ãƒ¼ãƒ é–‹å§‹ =====
# ãƒ—ãƒ¬ã‚¤ãƒ¤ãƒ¼ã¨ãƒ‡ã‚£ãƒ¼ãƒ©ãƒ¼ã«ã‚«ãƒ¼ãƒ‰ã‚’é…ã‚‹
player_hand = [deck.pop(), deck.pop()]
dealer_hand = [deck.pop(), deck.pop()]

# æ‰‹æœ­ã‚’è¡¨ç¤º
print("\nã‚ãªãŸã®æ‰‹æœ­:", player_hand, "â†’ åˆè¨ˆ:", calculate_score(player_hand))
print("ãƒ‡ã‚£ãƒ¼ãƒ©ãƒ¼ã®è¦‹ãˆã¦ã„ã‚‹ã‚«ãƒ¼ãƒ‰:", dealer_hand[0])


# ===== ãƒ—ãƒ¬ã‚¤ãƒ¤ãƒ¼ã®ã‚¿ãƒ¼ãƒ³ =====
while True:  #ã€Œï½“ã€ã‚’é¸æŠã™ã‚‹ã¾ã§ç¹°ã‚Šè¿”ã™
    print("\nã‚ãªãŸã®ç¾åœ¨ã®æ‰‹æœ­:", player_hand, "â†’ åˆè¨ˆ:", calculate_score(player_hand))
    choice = input("ã‚«ãƒ¼ãƒ‰ã‚’å¼•ãã¾ã™ã‹ï¼Ÿ (Hit=h / Stand=s): ")

    if choice == 'h':
        new_card = deck.pop()  #å±±æœ­ã‹ã‚‰æ–°ã—ã„ã‚«ãƒ¼ãƒ‰ã‚’å¼•ã
        player_hand.append(new_card)
        print(f"ã‚ãªãŸã¯ {new_card} ã‚’å¼•ãã¾ã—ãŸï¼")

        # ãƒãƒ¼ã‚¹ãƒˆãƒã‚§ãƒƒã‚¯
        if calculate_score(player_hand) > 21:
            print("ãƒãƒ¼ã‚¹ãƒˆã—ã¾ã—ãŸï¼ã‚ãªãŸã®è² ã‘ã§ã™ã€‚")
            break
    elif choice == 's':
        print("ã‚ãªãŸã¯ã‚«ãƒ¼ãƒ‰ã‚’å¼•ãã¾ã›ã‚“ã§ã—ãŸã€‚")
        break
    else:  #hã€sä»¥å¤–ã®æ–‡å­—ã‚’å…¥åŠ›ã—ãŸã¨ã
        print("h ã‹ s ã‚’å…¥åŠ›ã—ã¦ãã ã•ã„ã€‚")

# ===== ãƒ‡ã‚£ãƒ¼ãƒ©ãƒ¼ã®ã‚¿ãƒ¼ãƒ³ï¼ˆãƒ—ãƒ¬ã‚¤ãƒ¤ãƒ¼ãŒãƒãƒ¼ã‚¹ãƒˆã—ã¦ãªã‘ã‚Œã°ï¼‰=====
if calculate_score(player_hand) <= 21:  #ãƒ—ãƒ¬ã‚¤ãƒ¤ãƒ¼ãŒãƒãƒ¼ã‚¹ãƒˆã—ã¦ã„ãªã„ãªã‚‰
    print("\n--- ãƒ‡ã‚£ãƒ¼ãƒ©ãƒ¼ã®ã‚¿ãƒ¼ãƒ³ ---")
    print("ãƒ‡ã‚£ãƒ¼ãƒ©ãƒ¼ã®æ‰‹æœ­:", dealer_hand, "â†’ åˆè¨ˆ:", calculate_score(dealer_hand))

    player_score = calculate_score(player_hand)
    dealer_score = calculate_score(dealer_hand)

    # ãƒ‡ã‚£ãƒ¼ãƒ©ãƒ¼ã¯17æœªæº€ã¯å¿…ãšå¼•ã
    # ã•ã‚‰ã«18ã€œ20ã§ã‚‚ãƒ—ãƒ¬ã‚¤ãƒ¤ãƒ¼ã‚ˆã‚Šä½ã„ãªã‚‰å¼•ã
    while dealer_score < 17 or (17 <= dealer_score < 21 and dealer_score < player_score):  # ãƒ‡ã‚£ãƒ¼ãƒ©ãƒ¼ã¯17æœªæº€ã€ã‚‚ã—ãã¯18ã€œ20ã§ã‚‚ãƒ—ãƒ¬ã‚¤ãƒ¤ãƒ¼ã‚ˆã‚Šä½ã„ãªã‚‰å¿…ãšå¼•ã
        new_card = deck.pop()  #å±±æœ­ã‹ã‚‰æ–°ã—ã„ã‚«ãƒ¼ãƒ‰ã‚’å¼•ã
        dealer_hand.append(new_card)
        dealer_score = calculate_score(dealer_hand)
        print(f"ãƒ‡ã‚£ãƒ¼ãƒ©ãƒ¼ã¯ {new_card} ã‚’å¼•ãã¾ã—ãŸã€‚åˆè¨ˆ: {dealer_score}")

    # ===== å‹æ•—åˆ¤å®š ===== #
    print("\n--- å‹æ•—çµæœ ---")
    print("ã‚ãªãŸã®æ‰‹æœ­:", player_hand, "â†’ åˆè¨ˆ:", player_score)
    print("ãƒ‡ã‚£ãƒ¼ãƒ©ãƒ¼ã®æ‰‹æœ­:", dealer_hand, "â†’ åˆè¨ˆ:", dealer_score)

    if dealer_score > 21:
        print("ãƒ‡ã‚£ãƒ¼ãƒ©ãƒ¼ãŒãƒãƒ¼ã‚¹ãƒˆï¼ã‚ãªãŸã®å‹ã¡ã§ã™ğŸ‰")
    elif player_score > dealer_score:
        print("ã‚ãªãŸã®å‹ã¡ã§ã™ï¼ğŸ‰")
    elif player_score < dealer_score:
        print("ã‚ãªãŸã®è² ã‘ã§ã™â€¦ã€‚")
    else:
        print("å¼•ãåˆ†ã‘ã§ã™ã€‚")
else:
    print("\n--- ãƒ‡ã‚£ãƒ¼ãƒ©ãƒ¼ã®ã‚¿ãƒ¼ãƒ³ã¯ã‚¹ã‚­ãƒƒãƒ—ï¼ˆãƒ—ãƒ¬ã‚¤ãƒ¤ãƒ¼ãƒãƒ¼ã‚¹ãƒˆï¼‰ ---")
```
